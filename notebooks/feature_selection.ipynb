```python
import pandas as pd
import numpy as np 
from sklearn.feature_selection import mutual_info_classif
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("C:\dev\hybrid-phishing-detection\data\processed\clean_dataset.csv")
print(df.head())

```

```python
# understanding dataset
print ("dataset shape is : ", df.shape)
print("\n columns")
print(df.columns)
print("\ndataset information is: ", df.info)
```

```python
# putting all the input information in the x as input and all the output information in y as label 
x = df.drop("label", axis = 1)
y = df["label"]
x = x.select_dtypes(include=['number'])
# checking after removing string one kitne bche hai 
print("shape of dataset : ",x.shape)

```

```python
# doing mutual information 
mi_scores = mutual_info_classif(
    x,
    y,
    random_state=42
)
mi_df = pd.DataFrame({
    "Feature": x.columns,
    "MI Score": mi_scores
})
print(mi_df)
```

```python
# this is basically table ko sort kerne ka waay hai 

mi_df = mi_df.sort_values(
    by="MI Score",
    ascending=False
)
mi_top25 = mi_df.head(25)
print(mi_top25)
```

```python
dt = DecisionTreeClassifier(
    criterion="entropy",
    random_state=42
)
dt.fit(x,y)
ig_scores = dt.feature_importances_
ig_df = pd.DataFrame({
    "Feature": x.columns,
    "IG Score": ig_scores
})
ig_df = ig_df.sort_values(
    by = "IG Score", ascending= False
    )
ig_top25 = ig_df.head(25)
print(ig_top25)
```

```python
common_features = list(
    set(ig_top25["Feature"]).intersection(set(mi_top25["Feature"]))
)

print(common_features)
print("Number of common features:", len(common_features))
```

```python
comparison_df = pd.merge(
    ig_top25,
    mi_top25,
    on="Feature",
    how="inner",
    suffixes=("_IG", "_MI")
)

print(comparison_df)
```

```python
comparison_df.to_csv(
    r"C:\dev\hybrid-phishing-detection\data\processed\selected_features.csv",
    index=False
)
```
