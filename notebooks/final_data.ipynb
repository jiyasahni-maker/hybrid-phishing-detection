### Final data set production using the cleaned dataset and the selected features

```python
import pandas as pd

clean_df = pd.read_csv("../data/processed/clean_dataset.csv")

selected_df = pd.read_csv("../data/processed/selected_features.csv")
```

```python
selected_columns = selected_df["Feature"].tolist()

print(selected_columns)
```

```python
selected_columns.append("label")
```

```python
final_df = clean_df[selected_columns]
```

```python
print(final_df.shape)

final_df.head()
```

```python
final_df.to_csv(
    "../data/processed/final_dataset.csv",
    index=False
)
```
